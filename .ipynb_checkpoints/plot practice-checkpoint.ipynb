```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
def potential_energy(x, k=1):
    return 0.5*k*x**2

potential_energy(2,10)
```

```python
x= np.linspace(-10, 20, 100)
y= potential_energy(x)
print(np.transpose([x,y]))
```

```python
plt.plot(x,y, color="black")
```

```python
Contour plot
```

```python
def potential_energy_2D(x, y, kx=1.0, ky=1.0):
    return 0.5*(kx*x**2+ky*y**2)

potential_energy_2D(2,1)
```

```python
x= np.linspace(-5, 5, 10)
y= np.linspace(-5, 5, 10)

# combine 1D meshes into meshgrid
X, Y =np.meshgrid(x,y)
Z =potential_energy_2D(X,Y)
print(X)
print()
print(Y)
print()
print(Z)
```

```python
# construct the contour plot
plt.contour(X, Y, Z)

plt.colorbar(label="potential energy")

plt.xlabel("x-displacement")
plt.ylabel("y-displacement")

plt.show()

```

```python
# define the potential energy function
def p_energy(Q1, Q2, K1=1.0, K2=1.0, C=0):
    return 0.5*K1*Q1**2 + 0.5*K2*Q2**2 + C*Q1*Q2**2

# test p energy
p_energy(1,1)
```

```python
q1= np.linspace(-1,1,100)
q2= q1
Q1, Q2= np.meshgrid(q1, q2)

#p_energy(Q1, Q2)
```

```python
# plot contours with matplotlib contourf
plt.figure(figsize=(3,3))
plt.contour(Q1, Q2, p_energy(Q1, Q2, K1=10, K2=1, C=1), levels=25)
plt.colorbar(label='potential energy')
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.xlabel('$Q_1$')
plt.ylabel('$Q_2$')

plt.show()
```

```python
# define plotting function
def plot_p(K1=1, K2=2,C=0,n_contours=10, n_points=50):
    q1= np.linspace(-1,1,n_points)
    q2= q1
    Q1, Q2= np.meshgrid(q1, q2)
    Z= p_energy(Q1, Q2, K1=0.5, K2=1, C=-0.7)

    plt.figure(figsize=(3,3))
    plt.contour(Q1, Q2, Z, levels=25)
    plt.colorbar(label='potential energy')
    plt.xlim([-1,1])
    plt.ylim([-1,1])
    plt.xlabel('$Q_1$')
    plt.ylabel('$Q_2$')

    plt.show()
    return

```

```python
plot_p(K1=1, K2=2,C=0,n_contours=10, n_points=50)
```
